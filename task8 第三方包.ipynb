### 第三方包
Python有很多优秀的第三方包，这些包在安装python的时候没有安装，需要到pypi仓库里面去下载。你能想到，基本上都有。我们可以在命令行通过pip来安装第三方包。


```python
# 通过pip安装包
pip install 包名
# 通过pip卸载包
pip uninstall 包名
# 通过pip升级包
pip install --upgrade 包名
# 通过requirements.txt文本文件来安装依赖包
# requirements.txt文件中的数据如下
# 包名==版本
pip install -r requirements.txt
# 安装到用户目录
pip install --user 包名
# 安装指定版本
# ==号也可以为>,!=,<=等
pip install "包名==版本"
# 查看包的信息
pip show 包名
# 查找包的信息
pip search 包名
```

#### urllib3
urllib3是个功能强大的，用户友好的http客户端程序，应用广泛。urllib3实现了很多python官方没有实现的功能，比如线程安全、连接池、客户端SSL/TLS验证、压缩、代理等。

```python
#
import urllib3
http = urllib3.PoolManager()
r = http.request('GET','https://www.baidu.com')
print(r.status)
print(r.data)
```

#### Requests


```python
# requests包，http请求
# 这是个http访问的包
# 做爬虫的基础，如果真的做爬虫，建议使用scrapy
import requests
result = requests.get("https://www.njcit.cn")
if result.ok:
    print(result.content)
```

```python
# 带参数的请求
# 可以自己构造一些复杂的请求。
search = "哈哈"
response = requests.get("https://www.baidu.com/s?wd={}".format(search))
if response.ok:
    print(response.content)
```

```python
# Numpy包，数值计算
import numpy as np
array = np.arange(25).reshape(5,5)
print(array)
```

```python
#形状
print(array.shape)
#维度
print(array.ndim)
#数据类型
print(array.dtype.name)
#类型
print(type(array))
```

```python
# 矩阵加上常量
print(array+4)
```

```python
# 矩阵乘以常量
print(array*2)
```

```python
# 两个矩阵乘法运算
arrayb = np.array([2,3,4,5,6])
print(np.dot(array,arrayb))
```

```python
# 矩阵数据求和
print(array.sum())
```

```python
# 把数字拆成n等份
print(np.linspace(0, 100, 11))
```

```python
# 矩阵里面每个元素调用相同函数
print(np.sqrt(array))
```

```python
# SciPy 科学计算,有大量数学，科学，工程的相关算法
# 常量
from scipy import constants as const,linalg
import numpy as np
print(const.pi)
```

```python
# 求解方程组
#  x+2y=10
#  2x+y=11
a = np.array([[1,2],[2,1]])
b = np.array([10,11])
result = linalg.solve(a,b)
print(result)
```

#### Matplotlib
Matplotlib是增强的绘图库

```python
# pyplot 绘制更复杂的数学图形
#随机的散点图
import numpy as np
import matplotlib.pyplot as plt

N = 100
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (15 * np.random.rand(N))**2

plt.scatter(x,y,s=area,c=colors)
plt.show()
```

#### Flask
Flask是个轻量级的wsgi web应用框架，设计目标是简单和快速，也能满足复杂程序要求。具体学习可以参照中文网站https://dormousehole.readthedocs.io/en/latest/ 也可至官方网站 https://flask.palletsprojects.com/

```python
# Flask web框架,比较简单快速，如下是个最小的Flask应用代码，名称比如叫hello.py
#通过python flask.py运行后可以访问http://localhost:5000
from flask import Flask
app = Flask("任哥的Flash网站")

@app.route("/")
def index():
    return "欢迎访问任哥的网站"
```

- 在windows我们在命令行执行
    - set FLASK_APP=hello.py
- 然后我们执行
    - python -m flask run
- 最后我们访问http://localhost:5000

#### Django
Django是个基于python的web开发框架，功能强大，可以让你很容易的用很少的代码创建出更好的web应用。具体学习请参照https://www.django.cn/ 网站。官网：https://www.djangoproject.com/

- 安装Django，在命令行执行
    - python -m pip install Django
- 创建一个网站，在命令行执行
    - django-admin startproject mysite
- 系统会创建1一个mysite目录，里面有个manage.py和mysite目录，mysite目录下包含了基本的项目所需文件

- mysite/
    - manage.py
    - mysite/
        - __init__.py
        - settings.py
        - urls.py
        - asgi.py
        - wsgi.py
- 最后我们进入mysite目录执行
    - python manage.py runserver
- 最后我们打开http://127.0.0.1:8000 可以看到初始网络
